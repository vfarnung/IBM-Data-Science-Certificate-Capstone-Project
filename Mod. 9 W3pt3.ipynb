Import Relevant Libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

Get Text from Wikipedia Page and parse HTML code to get the table

```python
#get text from Wikipedia page
wikipedia_page ='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wikipedia_text = requests.get(wikipedia_page).text

# use beautiful soup to parse the HTML/XML codes.
soup = BeautifulSoup(wikipedia_text,'xml')
table = soup.find('table')
```

Transform table into a Dataframe, delete columns with unassigned Boroughs and Change name of unassigned neighbourhoods to name of borough

```python
data = []
columns = []

for index, tr in enumerate(table.find_all('tr')):
    section = []
    for td in tr.find_all(['th','td']):
        section.append(td.text.rstrip())
    
    #First row of data is the header
    if (index == 0):
        columns = section
    else:
        data.append(section)

#convert list into Pandas DataFrame
toronto_df = pd.DataFrame(data = data,columns = columns)
toronto_df = toronto_df[toronto_df['Borough'] != 'Not assigned']
toronto_df['Neighbourhood'].replace("Not assigned", toronto_df["Borough"],inplace=True)
toronto_df
```

```python
!wget -q -O 'toronto_data.csv' http://cocl.us/Geospatial_data
toronto_data=pd.read_csv('toronto_data.csv')
toronto_data.head()
```

```python
toronto_df_comb = pd.merge(toronto_df, toronto_data, how='left', left_on = 'Postal Code', right_on = 'Postal Code')
toronto_df_comb
```

Get geographical coordinates of Toronto from Geopy

```python
from geopy.geocoders import Nominatim 
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans

address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

Generate Toronto with Folium Map where circles are neighbourhoods and names of neighbourhoods are displayed when clicking onto the markers

```python
toronto_map = folium.Map(location=[latitude, longitude], zoom_start=13)

for lat, lng, borough, neighborhood in zip(toronto_df_comb['Latitude'], toronto_df_comb['Longitude'], toronto_df_comb['Borough'], toronto_df_comb['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.Circle(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto_map)  
    
toronto_map
```
